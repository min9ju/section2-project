```python
X_train = X_train.drop(["Android Ver","Price"], axis=1)
X_val = X_val.drop(["Android Ver","Price"], axis=1)
X_test = X_test.drop(["Android Ver","Price"], axis=1)
```

```python
# 결정트리 학습
# 과적합이 잘 일어날 수 있지만 위에서 제거해주었음

pipe_dt = make_pipeline(
    OrdinalEncoder(),
    DecisionTreeClassifier(random_state=42),
)
pipe_dt.fit(X_train, y_train)

pipe_dt_score = pipe_dt.score(X_val, y_val)
print("훈련 정확도", pipe_dt.score(X_train, y_train))
print("검증 정확도", pipe_dt_score)
```

```python
# 결정트리 학습
# 과적합이 잘 일어날 수 있지만 위에서 제거해주었음
# 또한 복잡도 줄이기 위해 max_dept 설정
# 검증 정확도 증가
pipe_dt = make_pipeline(
    OrdinalEncoder(),
    DecisionTreeClassifier(max_depth=8, random_state=42),
)
pipe_dt.fit(X_train, y_train)
pipe_dt_score = pipe_dt.score(X_val, y_val)
print("훈련 정확도", pipe_dt.score(X_train, y_train))
print("검증 정확도", pipe_dt_score)
```

```python
from sklearn.metrics import classification_report

y_pred = pipe_dt.predict(X_val)
print(classification_report(y_val, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_pred_proba = pipe_dt.predict_proba(X_val)[:, 1]
dt_auc = roc_auc_score(y_val, y_pred_proba)
print("AUC:", dt_auc)

baseline_fpr, baseline_tpr, _ = roc_curve(y_val, [0] * len(y_val))
fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)

plt.plot(baseline_fpr, baseline_tpr, linestyle="--", color="green", label="Baseline")
plt.plot(fpr, tpr, color="blue", label="Model")
plt.scatter(fpr, tpr, color="red")
plt.title("ROC curve")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```

```python
# 랜덤포레스트 모델 학습
# 결정트리기반, 하지만 이보다 상대적으로 과적합 해소 가능
# 노드 분할시 특성 n개중 일부분 k개의 특성을 랜덤하게 샘플링 최적 특성 찾아냄
from sklearn.ensemble import RandomForestClassifier

pipe_rf = make_pipeline(
    OrdinalEncoder(),
    RandomForestClassifier(max_depth=8, n_estimators=100, random_state=42, class_weight='balanced'),
)
pipe_rf.fit(X_train, y_train)
pipe_rf_score = pipe_rf.score(X_val, y_val)
print("훈련 정확도", pipe_rf.score(X_train, y_train))
print("검증 정확도", pipe_rf_score)
```

```python
from sklearn.metrics import classification_report

y_pred = pipe_rf.predict(X_val)
print(classification_report(y_val, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_pred_proba = pipe_rf.predict_proba(X_val)[:, 1]
rf_auc = roc_auc_score(y_val, y_pred_proba)
print("AUC:", rf_auc)

baseline_fpr, baseline_tpr, _ = roc_curve(y_val, [0] * len(y_val))
fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)

plt.plot(baseline_fpr, baseline_tpr, linestyle="--", color="green", label="Baseline")
plt.plot(fpr, tpr, color="blue", label="Model")
plt.scatter(fpr, tpr, color="red")
plt.title("ROC curve")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```

```python
# XGboost 사용
# 이전 모델이 틀린것을 직접 다시 학습해서 예측값을 순차적으로 보완

from category_encoders import OrdinalEncoder
from sklearn.metrics import classification_report, roc_auc_score
from xgboost import XGBClassifier

encoder = OrdinalEncoder()
X_train_encoded = encoder.fit_transform(X_train)  # 학습데이터
X_val_encoded = encoder.transform(X_val)  # 검증데이터

boosting = XGBClassifier(
    n_estimators=1000,
    objective="binary:logistic",  # default
    max_depth=6,
    learning_rate=0.2, # 학습률 : 잔차 어느정도 학습?
    subsample=0.8, # 과적합 막고 일반화 성능 올리기 위해 전체 데이터 중 일부를 샘플링하여 학습
    colsample_bytree=0.8 # 전체 칼럼중 일부를 샘플링하여 학습,
)

eval_set = [(X_train_encoded, y_train), (X_val_encoded, y_val)]

boosting.fit(X_train_encoded, y_train, eval_set=eval_set, early_stopping_rounds=50)
```

```python
y_pred_proba = boosting.predict_proba(
    X_val_encoded, iteration_range=(0, boosting.best_iteration)
)[:, 1]
y_pred = boosting.predict(X_val_encoded, iteration_range=(0, boosting.best_iteration))
boost_score = boosting.score(X_val_encoded, y_val)
print("검증 accuracy", boost_score)
print(classification_report(y_val, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score

y_pred_proba = boosting.predict_proba(
    X_val_encoded, iteration_range=(0, boosting.best_iteration)
)[:, 1]
boost_auc = roc_auc_score(y_val, y_pred_proba)
print("AUC:", boost_auc)

baseline_fpr, baseline_tpr, _ = roc_curve(y_val, [0] * len(y_val))
fpr, tpr, thresholds = roc_curve(y_val, y_pred_proba)

plt.plot(baseline_fpr, baseline_tpr, linestyle="--", color="green", label="Baseline")
plt.plot(fpr, tpr, color="blue", label="Model")
plt.scatter(fpr, tpr, color="red")
plt.title("ROC curve")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.show()
```
